```python
import pandas as pd
```

```python
t0 = pd.read_csv('ArchivoEntradaTrain0.csv', header=None)
t1 = pd.read_csv('ArchivoEntradaTrain1.csv', header=None)
t2 = pd.read_csv('ArchivoEntradaTrain2.csv', header=None)
t3 = pd.read_csv('ArchivoEntradaTrain3.csv', header=None)
t4 = pd.read_csv('ArchivoEntradaTrain4.csv', header=None)
t5 = pd.read_csv('ArchivoEntradaTrain5.csv', header=None)
t6 = pd.read_csv('ArchivoEntradaTrain6.csv', header=None)
t7 = pd.read_csv('ArchivoEntradaTrain7.csv', header=None)
t8 = pd.read_csv('ArchivoEntradaTrain8.csv', header=None)
t9 = pd.read_csv('ArchivoEntradaTrain9.csv', header=None)
```

```python
data_set = pd.concat([t0,t1,t2,t3,t4,t5,t6,t7,t8,t9],axis=0)
```

```python
data_set
```

```python
df_shuffled=data_set.sample(frac=1).reset_index(drop=True)
```

```python
df_shuffled
```

```python
df_shuffled.to_csv("data_train.csv",index=None,index_label=None,header=None)
```

```python
train_x = df_shuffled.iloc[:,:784]
train_y = df_shuffled.iloc[:,784:]
```

```python
train_x
```

```python
train_y
```

```python
train_x.to_csv('train_x.csv',index=None, index_label=None, header=None)
```

```python
train_y.to_csv('train_y.csv',index=None, index_label=None, header=None)
```
